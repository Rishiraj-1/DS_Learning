```python
import time 
import numpy as np

size = 1_000_000
list1 = list(range(size))
arr1 = np.array(list1)

start = time.time()
list_squares = [ x ** 2 for x in list1]
end = time.time()
print("python squaring time is :" ,end-start)

start = time.time()
numpy_squares = arr1 ** 2
end = time.time()
print("Numpy squaring time is :",end-start)
```
