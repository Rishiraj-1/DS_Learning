### vectorization 

```python
import numpy as np 
```

```python
arr = np.array([1,2,3,4,5])
```

```python
result = arr **2   #Vectorized operation 
```

```python
result
```

### Braodcasting


```python
result + 10  # here 10 is broadcasted along the whole array 
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
```

```python
arr2 = np.array([1,2,3])
```

```python
new = arr1 + arr2 #adding 1d array to an 2d array 
```

```python
new
```

here the 1d array is broadcasted along the whole 2d array 

```python
arr2 + arr1
```

### Simulating an data for normalized form 

```python
data = np.array([[10,20,30],[15,25,35],[20,30,40],[25,35,45],[30,40,50]])
```

```python
mean = data.mean(axis=0) 
```

```python
std = data.std(axis=0) 
```

```python
normalized_data = (data-mean)/std
```

```python
normalized_data
```
