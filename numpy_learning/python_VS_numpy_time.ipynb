### Program to Check speed of numpy and python

```python
import numpy as np
import time
# Python list
size = 1_000_000
list1 = list(range(size))
list2 = list(range(size))
start = time.time()
result = [x + y for x, y in zip(list1, list2)]
end = time.time()
print("Python list addition time:", end - start)
# NumPy array
arr1 = np.array(list1)
arr2 = np.array(list2)
start = time.time()
result = arr1 + arr2 # Vectorized operation
end = time.time()
print("NumPy array addition time:", end - start)
```
